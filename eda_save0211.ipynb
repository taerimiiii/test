# 탐색적 데이터 분석 EDA

- 데이터 시각화
    1. 상대방을 설득시키기 위한 자료를 만들 때
    2. 데이터를 빠르게 이해하고 파악할 때


- 단일변수 시각화
1) 연속형 : 해당 숫자데이터의 분포를 확인하는 목적
    - Histohram : 숫자데이터의 분포를 막대그래프로 표현 (X-연속형 / Y-빈도수)
    - Kernel Density Estimator (KDE, 확률 밀도) : 모수(모집단의 통계량)의 분포를 확률로 표현
                                                  (X-연속형 / Y-확률), 확률밀도함수와 비슷함
    - Box Plot (상자그림) : 숫자 데이터의 분포의 위치를 Box로 표현
                            (IQR : 전체 데이터의 50%가 분포해 있는 범위)
      
2) 범주형 : 해당 데이터의 빈도수를 확인하는 목적
    - 막대 그래프 / 파이차트

  
- 다변수 시각화
1) X 범주형 / Y 연속형 : 집단 간 통계량을 비교하는 경우
                -> 막대 그래프 (Bar Chart)
2) X 연속형 / Y 연속형 : 두 숫자 데이터의 상관성을 확인하는 경우 (비례 / 반비례)
               -> 산점도 (Scatter Plot)
3) X 순서형 / Y 연속형 : 시간/날짜 데이터에 따라 연속형 자료의 추세를 확인하는 경우
                -> 선그래프 (Line Chart)
4) 기타 : 관리도 (Control Chart) / 파레토도 (Pareto Plot) / ...

# 통계적 가설 검정

```python
import pandas as pd

# EUC-KR 인코딩으로 파일 읽기
df1 = pd.read_csv('02_Data.csv', encoding='EUC-KR')
```

```python
# 고객이 유입된 '채널' 별 '최종결제금액'의 평균의 차이가 있는지 확인!
df1.pivot_table(index='채널', value='최종결제금액', aggfunc='mean')
```

```python
# x - 범주형 / y - 연속형
# 'facebook'으로 유입된 고객들과, 'instagram'으로 유입된 고객들의
# '최종결제금액'의 평균의 차이가 있는가?
cond1 = df1['채널'] == 'facebook'
df1_face = df1.loc[cond1] # 'facebook'에 해당되는 고객 추출해 변수로 선언

cond2 = df1['채널'] == 'instagram'
df1_insta = df1.loc[cond2] # 'insta'에 해당되는 고객 추출해 변수로 선언
```

```python
# 가설 검정을 수행하는 Scipy 라이브러리 설치 (아나콘다 X)
# !pip --trusted-host pypi.org --trsuted-host files.pythonhost.org install scipy
```

```python
import scipy.stats as stats # 가설검정을 함수의 집합체 수행하는 라이브리리
```

```python
# 1-1. 각 집단의 연속형 자료의 정규성 검정
# 귀무가설 : 해당 숫자데이터는 정규분포를 따른다.
# 대립가설 : 해당 숫자데이터는 정규분포를 따르지 않는다.
stats.normaltest( df1_face['최종결제금액'] )
# P.value (귀무가설이 참일 확률) = 0%
# P.value < 5% (0.05) 유의수준 / 대립가설 참 / 숫자데이터는 정규분포를 따르지 않는다

# 실행결과
# NormaltestResult(statistic=19.080080970100784, pvalue=7.191393615906394e-05)
```

```python
# 1-2. (정규분포를 따르지 않는) 두 집단의 '최종결제금액'의 평균의 차이가 있는가?
# 귀무가설 : 두집단의 '최종결제금액' 평균의 차이는 없다.
# 대립가설 : 두 집단의 '최종결제금액' 평균의 차이는 있다.
stats.ranksums( df1_face['최종결제금액'], df1_insta['최종결제금액'] )
# P.value < 0.05 / 대립가설 참 / 두 집단의 '최종결제금액' 평균의 차이는 있다.

# 실행결과
# RanksumsResult(statistic=3.3310612893175584, pvalue=0.0008651555174077354)
```

```python
# 보조식품의 종류에 따라 결제상태가 달라지는가?
# '보조식품(추가구매)'와 '상태'가 서로 연관/독립인가?

# 귀무가설 : 두 데이터는 서로 독립이다.
#            보조식품의 종류에 따라 결제상태가 달라지지 않는다.
# 대립가설 : 두 데이터는 서로 독립이 아니다. (연관성이 있다.)
#            보조식품의 종류에 따라 결제상태가 달라진다.
```

```python
df1['보조식품(추가구매'].value_counts() #빈도수계산함수
```

```python
df1['상태'].value_counts()
```

```python
# crosstab( ) : 두 범주형 항목에 대한 빈도수를 계산
pd.crosstab( df1['보조식품(추가구매)'], df1['상태'] )
```

```python
# '상태'항목에서 cancelled 데이터를 제외한 고객들만 확인
cond1 = (df1['상태'] != 'cancelled')
df2 = df1.loc[cond1]
```

```python
df2_cross = pd.crosstab( df2['보조식품(추가구매)'], df2['상태'] ) # 표 같이 나옴
```

```python
stats.chi2_contingency(df2_cross)
# P.value 0.1294 > 0.05 (5%) / 귀무가설 참
# 두 데이터는 서로 독립이다. /
#            보조식품의 종류에 따라 결제상태가 달라지지 않는다.

# 실행결과
# Chil함수명(------------, pvalue=0.1293607007..., ------------)
```

```python
# x 연속형 / Y 연속형 (두 숫자가 서로 상관성이 있는가?
# '보조식품'을 구매한 고객들 중, '보조식품가격'과 '최종결제금액'의 상관성이 있는지?

# '보조식품'을 구매한 고객들만 데이터프레임 형태로 추출
cond1 = (df1['보조식품(추가구매)'] != '구매안함')
df3 = df1.loc[cond1]
```

```python
# 1-1. '보조식품가격'과 '최종결제금액'데이터가 정규분포를 띄는지? (정규성검정)
# 귀무가설 : 해당 숫자 데이터의 분포는 정규분포를 따른다.
# 대립가설 : 해당 숫자 데이터의 분포는 정규분포를 따르지 않는다.
stats.nvormaltest(df3['보조식품가격'])
# e-117 (10의 -119제곱, 0.000000...0000854..)
# P.value < 0.05 (유의수준 5%) / 대립가설 참 / 데이터의 분포는 정규분포를 따르지 않는다.

# 실행결과
# 함수명(--------------, pvalue=8.542506518439195e-118)
```

```python
stats.nvormaltest(df3['최종결제금액'])
# P.value < 0.05 / 대립가설 참 / 데이터의 분포는 정규분포를 따르지 않는다.

# 실행결과
# 함수명(--------------, pvalue=1.1018244793842086e-249)
```

```python
# 1-2. '보조식품가격'과 '최종결제금액'은 상관관계가 있는가?
# 귀무가설 : 두 연속형 데이터는 서로 상관성이 없다.
# 대립가설 : 두 연속형 데이터는 서로 상관성이 있다.
stats.spearmanr( df3['보조식품가격'], df3['최종결제금액'] )
# P.value < 0.05 / 대립가설 참 / 두 연속형 데이터는 서로 상관성이 있다.

# 실행결과
# 함수명(--------------, pvalue=4.3725233196527335e-262)
```
